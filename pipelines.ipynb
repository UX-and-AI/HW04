```python
from PIL import Image
from transformers import pipeline
```

### Image Description

New model definition/location:

```python
IMAGE_CAP_MODEL = "nlpconnect/vit-gpt2-image-captioning"
```

A test image:

```python
test_image = Image.open("./imgs/20.png").convert("RGB")
display(test_image)
```

The inference object:

```python
img_captioner = pipeline(task="image-to-text", model=IMAGE_CAP_MODEL)
```

Run inference:

```python
result = img_captioner(
    test_image, 
    generate_kwargs={
        "min_length": 20
    }
)
print(result[0]["generated_text"])
```
